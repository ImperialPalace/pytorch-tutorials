```python
import matplotlib.pyplot as plt
import numpy as np

# 生成一个二维随机数组
img = np.random.rand(5, 5)

print(img.shape)
print(img)

# 绘制灰度图像
plt.imshow(img, cmap='gray')

# 显示图像
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(precision=3)

# 生成一个随机的彩色图像
img = np.random.rand(5, 5, 3)

print(img.shape)

print(img)

# 绘制彩色图像
plt.imshow(img)

# 显示图像
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 生成一个二维随机数组
data = np.random.rand(10, 10)

# 绘制热力图
plt.imshow(data, cmap='hot')

# 显示图像
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# 加载地图图像, 下载地址：https://static.runoob.com/images/demo/map.jpeg
img = Image.open('images/map.jpeg')

# 转换为数组
data = np.array(img)

# 绘制地图
plt.imshow(data)

# 隐藏坐标轴
plt.axis('off')

# 显示图像
plt.show()
```
